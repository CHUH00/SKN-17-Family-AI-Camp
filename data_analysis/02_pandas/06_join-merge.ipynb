# DataFrame 병합

```python
import numpy as np
import pandas as pd
```

### merge

- 특정 '컬럼' 기준으로 DataFrame 병합

```python
df1 = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['다람쥐', '토끼', '곰', '고릴라']
})
df2 = pd.DataFrame({
    'id': [1, 2, 3, 5],
    'score': [100, 95, 88, 89]
})
```

```python
# merge()
# - on: merge 기준 컬럼
# - how: merge 방법 (inner(기본 값), left, right, outer)

pd.merge(df1, df2, on='id', how='inner')
```

```python
pd.merge(df1, df2, on='id', how='left')
```

```python
pd.merge(df1, df2, on='id', how='right')
```

```python
pd.merge(df1, df2, on='id', how='outer')
```

```python
df1 = pd.DataFrame({
    'n_id': [1, 2, 3, 4],
    'name': ['다람쥐', '토끼', '곰', '고릴라']
})
df2 = pd.DataFrame({
    's_id': [1, 2, 3, 5],
    'score': [100, 95, 88, 89]
})
```

```python
pd.merge(df1, df2, left_on='n_id', right_on='s_id')
```

### join

- 특정 '인덱스' 기준으로 DataFrame 병합

```python
df3 = pd.DataFrame({
    'age': [33, 48, 55, 24],
    'city': ['서울', '수원', '인천', '서울']
}, index=['이석훈', '김용준', '김진호', '남윤진'])
df4 = pd.DataFrame({
    'pet': ['고양이', '강아지', '햄스터', '돌']
}, index=['이석훈', '김용준', '김진호', '윤진'])

display(df3)
display(df4)
```

```python
df3.join(df4)       # how 기본값: left
```

```python
df3.join(df4, how='left')
```

```python
df3.join(df4, how='right')
```

```python
df3.join(df4, how='outer')
```

```python
df3.join(df4, how='inner')
```

### concat

- 축을 기준으로 DataFrame 병합

```python
df5 = pd.DataFrame({
    'name': ['구준표', '윤지후', '소이정', '프린스송'],
    'age': [18, 18, 18, 18]
})
df6 = pd.DataFrame({
    'name': ['금잔디'],
    'age': [17]
})
```

```python
pd.concat([df5, df6], axis=0).reset_index(drop=True)
```

```python
pd.concat([df5, df6], axis=1)
```

### melt

- 열 병합

```python
df = pd.DataFrame({
    'name': ['홍길동', '신사임당', '이순신'],
    'math': [80, 99, 100],
    'english': [34, 58, 48],
    'science': [99, 55, 77],
})

df
```

```python
pd.melt(
    df,                                         # 대상 DataFrame
    id_vars=['name'],                           # 고정(유지)될 컬럼
    value_vars=['math', 'english', 'science'],  # 합쳐질 컬럼
    var_name='subject',                         # 컬럼명이 합쳐질 컬럼의 이름
    value_name='score'                          # 컬럼값이 합쳐질 컬럼의 이름
)
```
