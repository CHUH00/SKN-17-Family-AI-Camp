```python
import json

with open('childcare_pages.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

preprocessed_data = []
for page in sorted(data.keys(), key=int):
    section_map = data[page]
    if not isinstance(section_map, dict) or not section_map:
        continue

    titles = list(section_map.keys())
    for title in titles:
        print(f"{title}")

        value = section_map[title]
        if isinstance(value, list):
            for line in value:
                preprocessed_data.append({"title": title, "content": line})
                print("-", line)
        else:
            print("-", value)
            preprocessed_data.append({"title": title, "content": line})
```
