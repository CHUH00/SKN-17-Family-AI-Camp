### 결측치 제거

```python
df_all = df_all[df_all['label'].str.strip() != ''].reset_index(drop=True)
```

```python
# 2. 각 pid 그룹 내에서 다음 연도의 label 값을 가져옵니다.
df_all['next_year_label'] = df_all.groupby('pid')['label'].shift(-1)
df_all
```

```python
# 3. 이탈(churn) 여부를 정의합니다.
# 조건 1: 현재 레이블이 2이고 (df_all['label'] == 2)
# 조건 2: 다음 해의 레이블이 존재하며 2가 아닌 경우 (df_all['next_year_label'].notna() & (df_all['next_year_label'] != 2))
# 위 두 조건을 만족하면 이탈(1), 아니면 비이탈(0)로 정의합니다.
churn_conditions = [
    (df_all['label'] == '2') & (df_all['next_year_label'].notna()) & (df_all['next_year_label'] != '2'),
    (df_all['label'] == '2') & (df_all['next_year_label'].notna()) & (df_all['next_year_label'] == '2')
]
churn_choices = [1, 0] # 1: 이탈 (Churn), 0: 비이탈 (Not Churn)

df_all['churn'] = np.select(churn_conditions, churn_choices, default=np.nan)
```

```python
# 4. 분석에 사용할 데이터셋을 생성합니다.
# 'churn' 값이 NaN이 아닌 경우, 즉 현재 label이 2인 고객들의 데이터만 필터링합니다.
# 이 데이터셋이 EDA와 모델링의 기반이 됩니다.
df_model_base = df_all[df_all['churn'].notna()].copy()

# 불필요해진 'next_year_label' 컬럼 삭제
df_model_base.drop(columns=['next_year_label'], inplace=True)

df_model_base['churn'] = df_model_base['churn'].astype(int)
```

```python
mapping = {
    '1': 11,  # 하루에도 여러 번 → 빈도 가장 높음
    '2': 10,  # 하루 1번(매일)
    '3': 9,   # 하루 5회 이상
    '4': 8,  # 하루 3~4회
    '5': 7,  # 하루 1~2회
    '6': 6,   # 1주일에 5~6회
    '7': 5,   # 1주일에 3~4회
    '8': 4,   # 1주일에 1~2회 정도
    '9': 3,   # 월 1~3회
    '10': 2,   # 월 1회 미만
    '11': 1    # 분기/연간 1회 정도 → 빈도 가장 낮음
}

df_model_base['d01'] = df_model_base['d01'].map(mapping).fillna(0).astype(int)
df_model_base['d02'] = df_model_base['d02'].map(mapping).fillna(0).astype(int)
df_model_base['d03'] = df_model_base['d03'].map(mapping).fillna(0).astype(int)
df_model_base['d04'] = df_model_base['d04'].map(mapping).fillna(0).astype(int)

df_model_base['relig'] = df_model_base['relig'].apply(lambda x: 0 if str(x).strip() == '' else x)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.preprocessing import StandardScaler

# 2. X, y 분리
X = df_model_base.drop(['churn'], axis=1)
y = df_model_base['churn']

# 3. 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# 4. 스케일링 (선택사항)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 5. 베이스 모델 정의 및 튜닝
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10],
}

param_grid_gb = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
}

rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5, scoring='roc_auc')
gb = GridSearchCV(GradientBoostingClassifier(random_state=42), param_grid_gb, cv=5, scoring='roc_auc')

# 6. 스태킹 모델 정의
estimators = [
    ('rf', rf),
    ('gb', gb),
    ('svc', SVC(probability=True, kernel='rbf', C=1.0, gamma='scale'))
]

stacking_clf = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(),
    cv=5,
    n_jobs=-1,
    passthrough=True
)

# 7. 학습
stacking_clf.fit(X_train_scaled, y_train)

# 8. 예측 및 평가
y_pred = stacking_clf.predict(X_test_scaled)
y_proba = stacking_clf.predict_proba(X_test_scaled)[:, 1]

print("Classification Report:\n", classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.preprocessing import StandardScaler

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

import warnings
warnings.filterwarnings('ignore', category=UserWarning)

# 1. 데이터 준비
X = df_model_base.drop(['churn'], axis=1)
y = df_model_base['churn']

# 2. 학습/테스트 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# 3. 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. 하이퍼파라미터 튜닝

# Random Forest
param_dist_rf = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [5, 10, 15, 20]
}
rf_search = RandomizedSearchCV(
    RandomForestClassifier(random_state=42),
    param_distributions=param_dist_rf,
    n_iter=10,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    random_state=42
)
rf_search.fit(X_train_scaled, y_train)
rf_best = rf_search.best_estimator_

# Gradient Boosting
param_dist_gb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}
gb_search = RandomizedSearchCV(
    GradientBoostingClassifier(random_state=42),
    param_distributions=param_dist_gb,
    n_iter=10,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    random_state=42
)
gb_search.fit(X_train_scaled, y_train)
gb_best = gb_search.best_estimator_

# XGBoost
param_dist_xgb = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.05, 0.1, 0.2]
}
xgb_search = RandomizedSearchCV(
    XGBClassifier(eval_metric='logloss', random_state=42),
    param_distributions=param_dist_xgb,
    n_iter=10,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    random_state=42
)
xgb_search.fit(X_train_scaled, y_train)
xgb_best = xgb_search.best_estimator_

# LightGBM
from sklearn.model_selection import GridSearchCV

param_grid_lgb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [5, 10, 15]
}

lgbm = GridSearchCV(
    estimator=LGBMClassifier(random_state=42, verbose=-1),
    param_grid=param_grid_lgb,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

lgbm.fit(X_train_scaled, y_train)
lgbm_best = lgbm.best_estimator_

# SVC (튜닝 없이 삽입)
svc = SVC(probability=True, kernel='rbf', C=1.0, gamma='scale')

# 5. 스태킹 모델 구성 (튜닝된 모델만 사용)
estimators = [
    ('rf', rf_best),
    ('gb', gb_best),
    ('xgb', xgb_best),
    ('lgbm', lgbm_best),
    ('svc', svc)
]

stacking_clf = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(max_iter=1000),
    cv=5,
    n_jobs=-1,
    passthrough=False
)

# 6. 학습
stacking_clf.fit(X_train_scaled, y_train)

# 7. 예측 및 평가
y_pred = stacking_clf.predict(X_test_scaled)
y_proba = stacking_clf.predict_proba(X_test_scaled)[:, 1]

print("📋 Classification Report:\n", classification_report(y_test, y_pred))
print("🎯 ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```
