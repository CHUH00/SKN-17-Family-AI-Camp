# Numpy 통계 관련

```python
import numpy as np
```

# 중심 경향 측도

```python
height = np.array([160, 170, 180, 172, 164, 152, 148, 192, 187, 177, 166, 152, 156, 176, 189])
```

```python
# 평균값
print(np.mean(height))
```

```python
# 중위값
print(np.median(height))
```

```python
print(np.sort(height))
```

```python
# 최빈값
values, count = np.unique(height, return_counts=True)
max_count = np.max(count)

print(values)
print(count == max_count)
print(values[count == max_count])
```

---

# 산포 측도

```python
# 분산 variance
# (값 - 평균)^2의 평균
var = np.mean(np.square(height - np.mean(height)))
print(var)

print(np.var(height))
```

```python
# 포준편차 standard deviation
# 분산의 제곱근

std = np.sqrt(var)
print(std)

print(np.std(height))
```

```python
# 범위
# 25% 지점 : 1사분위 값
# 50% 지점 : 2사분위 값 == 중위값
# 75% 지점 : 3사분위 값
# 100% 지점 : 4사분위 값 == 최대값

print(np.max(height) - np.min(height))
```

# 확률 분포

```python
# !pip install matplotlib

import matplotlib.pyplot as plt
```

```python
# 표준 정규 분포: 평균이 0, 표준편차가 1: 종 모양의 분포

# np.random.randn(shape)
np.random.randn()
```

```python
arr = np.random.randn(1000)   # (1000,)

plt.hist(arr, bins=100)
plt.show()
```

```python
# 정규 분포: 평균이 0, 표준편차가 1: 종 모양의 분포

# np.random.normal(loc, scale, size)
# - loc: 평균
# - scale: 표준편차
# - size: 갯수 or shape
arr = np.random.normal(10, 1, 100)          # 갯수
print(arr.shape)

plt.hist(arr, bins=100)
plt.show()

arr = np.random.normal(10, 1, (10, 10))     # shape
print(arr.shape)

plt.hist(arr, bins=10)
plt.show()
```

```python
# 균등분포 : 모든 값이 동일한 값으로 비등비등하게 존재 / 일자 형태

# np.random.rand(num)
# - 0.0 이상 1.0 미만 수에서 균등하게 난수 추출
# - num: 개수 or shape
arr = np.random.rand(1000)
print(arr)
print(np.max(arr), np.min(arr))

plt.hist(arr, bins=10)
plt.show()
```

```python
# 균등분포 : 모든 값이 동일한 값으로 비등비등하게 존재 / 일자 형태

# np.random.randint(start, end, size)
# start 이상 end 미만의 난수(정수)를 size(shape)만큼 추출
np.random.seed(0)   # 난수 계산 시작값 seed 고정

arr = np.random.randint(1, 100, 1000)
arr = np.random.randint(1, 100, (2, 3))

print(arr)
print(np.max(arr), np.min(arr))

plt.hist(arr, bins=100)
plt.show()
```
